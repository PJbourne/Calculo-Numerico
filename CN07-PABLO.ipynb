```python
#1

import math as m

def difRegrProg(f, x, h):
    return ( f(x+h) - f(x) ) / h
def tresDotMed(f, x, h):
    return ( f(x+h)-f(x-h) ) / (2*h)
def cincoDotMed(f, x, h):
    return (f(x-2*h) - 8*f(x-h) + 8*f(x+h) - f(x + 2*h)) / (12 * h)
```

```python
#2
def a(x):
    return m.exp(2*x)
def b(x):
    return x*m.log(x)
def c(x):
    return x*m.cos(x) - x**2 * m.sin(x)
def d(x):
    return m.log(x + 2) - (x + 1)**2
def e(x):
    return (m.cos(3*x))**2 - m.exp(2*x)
def f(x):
    return (2*x) / (m.cos(x) + 3)

h1 = 0.1
h2 = 0.01
x = 2.1

funcoes = [a,b,c,d,e,f]
results = []
erros = []

for i in range(len(funcoes)):
    results.append(difRegrProg(funcoes[i], x, h1))
    results.append(difRegrProg(funcoes[i], x, h2))
    results.append(tresDotMed(funcoes[i], x, h1))
    results.append(tresDotMed(funcoes[i], x, h2))
    results.append(cincoDotMed(funcoes[i], x, h1))
    results.append(cincoDotMed(funcoes[i], x, h2))
    erros.append(abs( 133.372662081 - results[i] ))
    erros.append(abs( 1.74193734472938 - results[i] ))
    erros.append(abs( -3.63856518763652 - results[i] ))
    erros.append(abs( -5.956097956097956097 - results[i] ))
    erros.append(abs( -134.02709180604 - results[i] ))
    erros.append(abs( 0.50970616930589 - results[i] ))
    #results = results.append(difRegrProg(funcoes[i], x, h1))

for i in range(36):
    print(f'Cálculo de derivada para {x}, com intervalos {h1} e {h2}.\n')
    print(f'Dif. Regr/Progr.\n \tPara h1: {results[i]}. Erro = {erros[i]}\n \tPara h2: {results[i]}. Erro = {erros[i]}')
    print(f'F. 3 pontos;\n \tPara h1: {results[i]}. Erro = {erros[i]}\n \tPara h2: {results[i]}. Erro = {erros[i]}')
    print(f'F. 5 pontos;\n \tPara h1: {results[i]}. Erro = {erros[i]}\n \tPara h2: {results[i]}. Erro = {erros[i]}')


    
```
