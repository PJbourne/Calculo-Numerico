```python
#1.1

import math as m
import pandas as pd

def difRegrProg(f, x, h):
    return ( f(x+h) - f(x) ) / h
def tresDotMed(f, x, h):
    return ( f(x+h)-f(x-h) ) / (2*h)
def cincoDotMed(f, x, h):
    return (f(x-2*h) - 8*f(x-h) + 8*f(x+h) - f(x + 2*h)) / (12 * h)
```

```python
#1.2
def a(x):
    return m.exp(2*x)
def b(x):
    return x*m.log(x)
def c(x):
    return x*m.cos(x) - x**2 * m.sin(x)
def d(x):
    return m.log(x + 2) - (x + 1)**2
def e(x):
    return (m.cos(3*x))**2 - m.exp(2*x)
def f(x):
    return (2*x) / (m.cos(x) + 3)

h1 = 0.1
h2 = 0.01
x = 2.1

funcoes = [a,b,c,d,e,f]
results = []
erros = []
real = [133.372662081, 1.74193734472938, -3.71669379320239, -5.95609756097561, -133.473531223514, 1.38388590556836]
#matrixResults = pd.DataFrame(6,6)

for i in range(6):
    results.append(difRegrProg(funcoes[i], x, h1))
    results.append(difRegrProg(funcoes[i], x, h2))
    results.append(tresDotMed(funcoes[i], x, h1))
    results.append(tresDotMed(funcoes[i], x, h2))
    results.append(cincoDotMed(funcoes[i], x, h1))
    results.append(cincoDotMed(funcoes[i], x, h2))

for i in range(6):
    erros.append(abs( real[i] - results[i*6] ))
    erros.append(abs( real[i] - results[i*6+1] ))
    erros.append(abs( real[i] - results[i*6+2] ))
    erros.append(abs( real[i] - results[i*6+3] ))
    erros.append(abs( real[i] - results[i*6+4] ))
    erros.append(abs( real[i] - results[i*6+5] ))

for i in range(6):
    print(f'Cálculo de derivada da equação {i+1} para x = {x}, com intervalos h1={h1} e h2={h2}.\n')
    print(f'Dif. Reg/Pr.\tPara h1: {results[i*6]:.5}. E = {erros[i*6]:.5}\n \t\tPara h2: {results[i*6+1]:.5}. E = {erros[i*6+1]:.5}')
    print(f'F. 3 pts;\tPara h1: {results[i*6+2]:.5}. E = {erros[i*6+2]:.5}\n \t\tPara h2: {results[i*6+3]:.5}. E = {erros[i*6+3]:.5}')
    print(f'F. 5 pts;\tPara h1: {results[i*6+4]:.5}. E = {erros[i*6+4]:.5}\n \t\tPara h2: {results[i*6+5]:.5}. E = {erros[i*6+5]:.5}')
```

```python
#2.1

def trap(f, a, b):
    x0 = a
    x1 = b
    h = b - a
    return h*(f(x0)+f(x1)) / 2

def simp(f, a, b):
    h = (b - a) / 2
    x0 = a
    x1 = a + h
    x2 = b
    return h*(f(x0) + 4*f(x1) + f(x2)) / 3

def simP(f, a, b):
    h = (b - a) / 3
    x0 = a
    x1 = a + h
    x2 = a + 2*h
    x3 = b
    return 3*h*(f(x0) + 3*f(x1) + 3*f(x2) + f(x3)) / 8
```

```python
#2.2
a = 0
b = 2

def A(x):
    return x**2

def B(x):
    return x**4

def C(x):
    return 1 / (x + 1)

def D(x):
    return m.sqrt(x + 1)

def E(x):
    return m.sin(x)

def F(x):
    return m.exp(x)

funcoes2 = [A,B,C,D,E,F]
reaL = [2.6666666666666666, 6.4, 1.09861228866811, 2.78743494847109, 1.41614683654714, 6.38905609893065]
res = []
err = []
for i in range(len(funcoes2)):
    res.append(trap(funcoes2[i], a, b))
    res.append(simp(funcoes2[i], a, b))
    res.append(simP(funcoes2[i], a, b))

for i in range(len(funcoes2)):
    err.append(abs(reaL[i] - res[i*3]))
    err.append(abs(reaL[i] - res[i*3+1]))
    err.append(abs(reaL[i] - res[i*3+2]))

for i in range(len(funcoes2)):
    print(f'Integral da equação {i+1}; intervalos: a={a} b={b}')
    print(f'\tR: {res[i*3]:.6}. \tE:{err[i*3]:.6}')
    print(f'\tR: {res[i*3+1]:.6}. \tE: {err[i*3+1]:.6}')
    print(f'\tR: {res[i*3+2]:.6}. \tE: {err[i*3+2]:.6}')
```
